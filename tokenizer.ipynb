```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(binary=True)
corpus = ["Tessaract is good optical character recognition engine  ", "optical character recognition is significant "]
vect.fit(corpus)
print(vect.transform(["This is a good optical illusion"]).toarray())
```

```python
print(vect.transform(corpus).toarray())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vect.transform(["Google Cloud Vision is a character recognition engine"]).toarray(), vect.transform(["OCR is an optical character recognition engine"]).toarray())
```

# frequency distribution of words in a text

```python
from nltk.probability import ConditionalFreqDist
cfd=ConditionalFreqDist((len(word),word) for word in text.split())
#list of conditons
cfd.conditions()
```

```python
inauguraldata = inaugural.words(fileids = '2009-Obama.txt')
cdf = ConditionalFreqDist((len(word), word) for word in inauguraldata)
cdf[4]
```

```python
data = []
for words in inauguraldata:
    if len(words)>4:
        data.append(words)
```

```python
fd = nltk.FreqDist(data)
fd
```

```python
sorted_fd = sorted(fd.items(), key = lambda x:x[1])
```

```python
word = list(sorted_fd)[len(sorted_fd)-1]
```
