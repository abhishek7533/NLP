```python
from nltk.stem.porter import PorterStemmer
st = PorterStemmer()
df= text4[:1500]
for word in df:
    print(word,st.stem(word))
```

```python
from nltk.stem import WordNetLemmatizer
wn = WordNetLemmatizer()
for word in df:
    print(word,wn.lemmatize(word))
```

```python
from nltk.tokenize import word_tokenize
```

```python
for word in df:
    print(word,word_tokenize(word))
```

# #CONDITIONAL FREQUENCY DISTRIBUTION

```python
from nltk.probability import ConditionalFreqDist
cfd=ConditionalFreqDist((len(word),word) for word in text1.split())
cfd[3]
```

# frequency distribution of words in a text

```python
from nltk.probability import ConditionalFreqDist
cfd=ConditionalFreqDist((len(word),word) for word in text.split())
#list of conditons
cfd.conditions()
```

```python
cfd[8]
```

```python
inauguraldata = inaugural.words(fileids = '2009-Obama.txt')
cdf = ConditionalFreqDist((len(word), word) for word in inauguraldata)
cdf[4]
```

```python
data = []
for words in inauguraldata:
    if len(words)>4:
        data.append(words)
```

```python
fd = nltk.FreqDist(data)
fd
```

```python
sorted_fd = sorted(fd.items(), key = lambda x:x[1])
```

```python
word = list(sorted_fd)[len(sorted_fd)-1]
```

```python
print("Frequently Used Word is :",word)
```
