# Using NLTK for Indian Languages 

```python
import nltk
```

```python
from nltk.corpus import indian
from nltk.tag import tnt
```

```python
indian.fileids()
```

```python
# To find number of characters in each language corpora

for i in indian.fileids():
    print(i, len(indian.raw(i)))
```

```python
# To find number of words in each language corpus

for i in indian.fileids():
    print(i, len(indian.words(i)))
```

```python
#To find number of sentences

for i in indian.fileids():
    print(i, len(indian.sents(i)))
```

## Hindi Corpus

```python
# accessing sentences from hindi corpus
hindi_sent = indian.sents('hindi.pos')
```

```python
for h_sent in hindi_sent:
    print(h_sent)
```

```python
# storing these sentences in a file
it = open("hindi_corpus_sents.txt",'w',encoding="utf-8")  # creating a file
for h_sent in hindi_sent:
    it.write(" ".join(h_sent))  # will convert the list of words in to actual sentence and will write all sentences to the file
```

```python
# accessing the words in corpus
hin_word = indian.words('hindi.pos')
hin_word
```

```python
# accessing the POS tagged sentences from a corpora
hindi_pos_sen = indian.tagged_sents('hindi.pos')
for sent in hindi_pos_sen:
    print(sent)
```
